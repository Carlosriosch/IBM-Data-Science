```python
import matplotlib.pyplot as plt
```

```python
y=[]
x=[]
r=3
y.append(0.4)
x.append(0)
for i in range(35):
    yn=y[-1]
    yy=r*yn*(1-yn)
    y.append(yy)
    x.append(i+1)

```

```python
plt.plot(x,y)

```
